# Day 18 - Plotting with Matplotlib: วาดกราฟพื้นฐาน
จุดประสงค์: เรียนรู้การวาดกราฟเส้นและการปรับแต่งด้วย Matplotlib

1. การนำเข้า Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

2. การสร้างข้อมูลตัวอย่าง

```python
x = np.linspace(0, 10, 100)  # 0 ถึง 10 มี 100 จุด
y = np.sin(x)
```

3. การวาดกราฟเส้นพื้นฐาน

```python
plt.plot(x, y)
plt.show()
```

4. การใส่ชื่อกราฟและชื่อแกน

```python
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis (time)")
plt.ylabel("Y-axis (amplitude)")
plt.grid(True)
plt.show()
```

5. การปรับแต่งเส้นและ marker

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', markersize=4)
plt.title("Sine Wave with Style")
plt.xlabel("X-axis")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

6. การวาดหลายเส้นในกราฟเดียว

```python
y2 = np.cos(x)
```

```python
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title("Sine and Cosine Waves")
plt.xlabel("X-axis")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

7. การบันทึกกราฟเป็นไฟล์

```python
plt.plot(x, y)
plt.title("Sine Wave")
plt.savefig("sine_wave.png", dpi=300)
```
