# Day 17 - Generating Data: linspace, arange
จุดประสงค์: เรียนรู้การสร้างชุดข้อมูลตัวเลขด้วย np.arange() และ np.linspace()

**1. การใช้ `np.arange(start, stop, step)`**

- สร้างชุดตัวเลขจาก `start` ถึงก่อน `stop` โดยเพิ่มทีละ `step`

```python
import numpy as np

a = np.arange(0, 10, 2)  # 0, 2, 4, 6, 8
```

```python
print(a)
```

**2. การใช้ `np.linspace(start, stop, num)`**

- สร้างชุดตัวเลขจาก `start` ถึง `stop` โดยมีจำนวนจุดเท่ากับ `num`

```python
b = np.linspace(0, 1, 5)  # 0.00, 0.25, 0.50, 0.75, 1.00
```

```python
print(b)
```

**3. ความแตกต่างหลัก**

| ฟังก์ชัน | กำหนดอะไร | เหมาะกับกรณี |
| --- | --- | --- |
| `arange()` | start, stop, step | รู้ระยะห่างที่ต้องการ |
| `linspace()` | start, stop, จำนวนจุด | รู้จำนวนจุดที่ต้องการ |

**4. ตัวอย่างการใช้งานจริง**

- สร้างข้อมูลเวลาเพื่อ plot กราฟ sine:

```python
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 100) # 100 จุด
y = np.sin(t)
```

```python
plt.plot(t, y)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Sine Wave")
plt.show()
```

5. ใช้ร่วมกับการจำลอง

```python
# สร้างค่าความเร็วจาก 0 ถึง 100 m/s เพิ่มทีละ 5
speeds = np.arange(0, 105, 5)
```

```python
print("Speeds:", speeds)
```

```python
# สร้างค่ามุมจาก 0 ถึง 90 องศา แบ่งเป็น 10 จุด
angles = np.linspace(0, 90, 10)
```

```python
print("Angles:", angles)
```
