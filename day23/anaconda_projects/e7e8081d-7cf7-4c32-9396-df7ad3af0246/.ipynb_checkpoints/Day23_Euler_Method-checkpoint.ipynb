## หลักการของ Euler’s Method

ถ้าเรามีสมการเชิงอนุพันธ์:

$$
\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0
$$

Euler’s Method คำนวณโดย:

โดย

$$
y_{n+1} = y_n + h \cdot f(t_n, y_n)
$$

- $h$ = ขนาดขั้นเวลา (time step)
- $t_n$ = เวลาปัจจุบัน
- $y_n$ = ค่าปัจจุบันของตัวแปร

**โจทย์:** $\frac{dy}{dt} = -2y$, $y(0) = 5$, แก้ในช่วง $t = 0$ ถึง $t = 5$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ฟังก์ชัน f(t, y)
def f(t, y):
    return -2 * y
```

```python
# เงื่อนไขเริ่มต้น
y0 = 5
t0 = 0
t_end = 5
h = 0.1  # ขนาดขั้นเวลา
```

```python
# เตรียม array เก็บค่า
t_values = np.arange(t0, t_end + h, h)
y_values = np.zeros(len(t_values))
```

```python
# กำหนดค่าเริ่มต้น
y_values[0] = y0
```

```python
# Euler Method
for i in range(len(t_values) - 1):
    y_values[i+1] = y_values[i] + h * f(t_values[i], y_values[i])
```

```python
# คำตอบเชิงวิเคราะห์
y_exact = y0 * np.exp(-2 * t_values)
```

```python
# Plot
plt.plot(t_values, y_values, 'o-', label='Euler Method')
plt.plot(t_values, y_exact, 'r', label='Exact Solution')
plt.xlabel('Time t')
plt.ylabel('y(t)')
plt.title('Euler Method vs Exact Solution')
plt.legend()
plt.grid(True)
plt.show()
```
