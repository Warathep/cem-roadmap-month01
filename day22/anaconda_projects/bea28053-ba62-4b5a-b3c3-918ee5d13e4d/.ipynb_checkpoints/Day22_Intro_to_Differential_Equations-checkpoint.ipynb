### 1. สมการเชิงอนุพันธ์คืออะไร

- เป็นสมการที่มี **อนุพันธ์ (derivative)** ของตัวแปรหนึ่งหรือมากกว่า
- ตัวอย่าง:
    
    $$
    \frac{dy}{dt} = ky
    $$
    
    หมายถึง อัตราการเปลี่ยนแปลงของ (y) แปรผันตรงกับค่า (y) เอง

### 2. ตัวอย่างในวิศวกรรม

- **การสลายตัวของสารกัมมันตรังสี**: $\frac{dN}{dt} = -\lambda N$
- **การชาร์จ/คายประจุของตัวเก็บประจุ (RC circuit)**: $\frac{dV}{dt} = -\frac{V}{RC}$
- **การเคลื่อนที่แบบโปรเจกไทล์**: $\frac{dv}{dt} = -g$

### 3. การแก้สมการเชิงอนุพันธ์อย่างง่ายด้วย SciPy

ตัวอย่าง: $\frac{dy}{dt} = -2y, \space y(0) = 5$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# นิยามสมการ dy/dt = -2y
def model(y, t):
    dydt = -2 * y
    return dydt
```

```python
# เงื่อนไขเริ่มต้น
y0 = 5
```

```python
# ช่วงเวลา
t = np.linspace(0, 5, 100)
```

```python
# แก้สมการ
y = odeint(model, y0, t)
```

```python
# Plot
plt.plot(t, y)
plt.xlabel('Time')
plt.ylabel('y(t)')
plt.title('Solution of dy/dt = -2y')
plt.grid(True)
plt.show()
```
