```python
def simulate_ode(f, y0, t_start, t_end, h):
    """
    ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏à‡∏≥‡∏•‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÅ‡∏Å‡πâ ODE ‡∏î‡πâ‡∏ß‡∏¢ Euler Method
    f: ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏ô‡∏¥‡∏¢‡∏≤‡∏° dy/dt = f(t, y)
    y0: ‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô
    t_start, t_end: ‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏ß‡∏•‡∏≤
    h: step size
    """
    t_values = np.arange(t_start, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0

    for i in range(len(t_values) - 1):
        y_values[i+1] = y_values[i] + h * f(t_values[i], y_values[i])

    return t_values, y_values
```

**‡∏™‡∏°‡∏Å‡∏≤‡∏£:** $\frac{dy}{dt} = -2y$, $y(0) = 5$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ‡∏ô‡∏¥‡∏¢‡∏≤‡∏°‡∏™‡∏°‡∏Å‡∏≤‡∏£
def decay(t, y):
    return -2 * y
```

```python
# ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡πÉ‡∏ä‡πâ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏à‡∏≥‡∏•‡∏≠‡∏á
t, y = simulate_ode(decay, y0=5, t_start=0, t_end=5, h=0.1)
```

```python
# ‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏à‡∏£‡∏¥‡∏á
y_exact = 5 * np.exp(-2 * t)
```

```python
# Plot
plt.plot(t, y, 'o-', label='Euler Simulation')
plt.plot(t, y_exact, 'r--', label='Exact Solution')
plt.xlabel("Time")
plt.ylabel("y(t)")
plt.title("General Simulation Function Example")
plt.legend()
plt.grid(True)
plt.show()
```
