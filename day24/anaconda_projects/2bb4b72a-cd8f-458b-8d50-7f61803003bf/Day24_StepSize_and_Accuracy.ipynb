### 1. ‡∏™‡∏°‡∏Å‡∏≤‡∏£‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡∏ó‡∏î‡∏™‡∏≠‡∏ö

$$
\frac{dy}{dt} = -2y, \quad y(0) = 5
$$

‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏à‡∏£‡∏¥‡∏á:

$$
y(t) = 5e^{-2t}
$$

2. ‡πÇ‡∏Ñ‡πâ‡∏î‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö step size

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t, y):
    return -2 * y
```

```python
def euler_method(f, y0, t0, t_end, h):
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0
    for i in range(len(t_values) - 1):
        y_values[i+1] = y_values[i] + h * f(t_values[i], y_values[i])
    return t_values, y_values
```

```python
# ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô
y0 = 5
t0 = 0
t_end = 5
```

```python
# ‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏à‡∏£‡∏¥‡∏á
t_exact = np.linspace(t0, t_end, 1000)
y_exact = y0 * np.exp(-2 * t_exact)
```

```python
# ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö step size
step_sizes = [0.5, 0.2, 0.05]
```

```python
plt.plot(t_exact, y_exact, 'k-', label='Exact Solution')

plt.title("Effect of Step Size on Accuracy")
plt.xlabel("Time t")
plt.ylabel("y(t)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
for h in step_sizes:
    t_num, y_num = euler_method(f, y0, t0, t_end, h)
    plt.plot(t_num, y_num, marker='o', label=f'Euler h={h}')

plt.title("Effect of Step Size on Accuracy")
plt.xlabel("Time t")
plt.ylabel("y(t)")
plt.legend()
plt.grid(True)
plt.show()
```
