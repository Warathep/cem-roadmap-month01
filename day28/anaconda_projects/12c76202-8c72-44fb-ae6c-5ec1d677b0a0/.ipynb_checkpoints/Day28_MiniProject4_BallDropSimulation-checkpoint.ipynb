## üìö ‡∏´‡∏•‡∏±‡∏Å‡∏Å‡∏≤‡∏£‡∏ü‡∏¥‡∏™‡∏¥‡∏Å‡∏™‡πå‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ

- ‡∏™‡∏°‡∏°‡∏ï‡∏¥‡∏ß‡πà‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÅ‡∏£‡∏á‡∏ï‡πâ‡∏≤‡∏ô‡∏≠‡∏≤‡∏Å‡∏≤‡∏®
- ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πà‡∏á‡∏à‡∏≤‡∏Å‡πÅ‡∏£‡∏á‡πÇ‡∏ô‡πâ‡∏°‡∏ñ‡πà‡∏ß‡∏á: $g = 9.81 \space \text{m/s}^2$
- ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏π‡∏á:
    
    $$
    h(t) = h_0 - \frac{1}{2}gt^2
    $$
    
- ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß:
    
    $$
    v(t) = -gt
    $$

üñ•Ô∏è ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡πÇ‡∏Ñ‡πâ‡∏î‡∏à‡∏≥‡∏•‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏Å‡∏Ç‡∏≠‡∏á‡∏•‡∏π‡∏Å‡∏ö‡∏≠‡∏•

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ‡∏û‡∏≤‡∏£‡∏≤‡∏°‡∏¥‡πÄ‡∏ï‡∏≠‡∏£‡πå
h0 = 20       # ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏π‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô (‡πÄ‡∏°‡∏ï‡∏£)
g = 9.81      # ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πà‡∏á‡πÇ‡∏ô‡πâ‡∏°‡∏ñ‡πà‡∏ß‡∏á (m/s^2)
t_end = np.sqrt(2 * h0 / g)  # ‡πÄ‡∏ß‡∏•‡∏≤‡∏ó‡∏µ‡πà‡∏•‡∏π‡∏Å‡∏ö‡∏≠‡∏•‡∏ñ‡∏∂‡∏á‡∏û‡∏∑‡πâ‡∏ô
t = np.linspace(0, t_end, 100)
```

```python
# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡πÅ‡∏•‡∏∞‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß
height = h0 - 0.5 * g * t**2
velocity = -g * t
```

```python
# Plot ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏π‡∏á
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(t, height, label='Height (m)', color='blue')
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.title("Ball Drop: Height vs Time")
plt.grid(True)
plt.legend()
```

```python
# Plot ‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏£‡πá‡∏ß
plt.subplot(1, 2, 2)
plt.plot(t, velocity, label='Velocity (m/s)', color='red')
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Ball Drop: Velocity vs Time")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
