# Day 19 - Multiple Plots: เปรียบเทียบหลายเส้นในกราฟเดียว
จุดประสงค์: เรียนรู้การวาดหลายเส้นในกราฟเดียวและการใช้ legend

1. การสร้างข้อมูลตัวอย่าง

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)
```

2. การวาดหลายเส้นในกราฟเดียว

```python
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='red')
plt.plot(x, y3, label='sin(x) + cos(x)', color='green')

plt.title("Multiple Plots Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # แสดง legend
plt.grid(True)
plt.show()
```

3. การปรับแต่งสไตล์แต่ละเส้น

```python
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', marker='o', markersize=4)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', marker='s', markersize=4)
plt.plot(x, y3, label='sin(x) + cos(x)', color='green', linestyle='-.', marker='^', markersize=4)

plt.title("Multiple Plots with Styles")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

4. การบันทึกกราฟ

```python
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
plt.savefig("multiple_plots.png", dpi=300)
```
