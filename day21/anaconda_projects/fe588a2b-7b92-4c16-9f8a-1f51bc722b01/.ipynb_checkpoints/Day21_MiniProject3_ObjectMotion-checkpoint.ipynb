## หลักการฟิสิกส์ที่ใช้

สมมติว่าไม่มีแรงต้านอากาศ

- **สมการการเคลื่อนที่แนวนอน**
    
    $$
    x(t) = v_0 \cos(\theta) \cdot t
    $$
    
- **สมการการเคลื่อนที่แนวตั้ง**
    
    $$
    y(t) = v_0 \sin(\theta) \cdot t - \frac{1}{2} g t^2
    $$
    

โดย

- $v_0$ = ความเร็วเริ่มต้น (m/s)
- $\theta$ = มุมยิง (องศา)
- $g$ = ความเร่งเนื่องจากแรงโน้มถ่วง (9.81 m/s²)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ค่าพารามิเตอร์
v0 = 20       # ความเร็วเริ่มต้น (m/s)
theta = 45    # มุมยิง (องศา)
g = 9.81      # ความเร่งโน้มถ่วง (m/s^2)
```

```python
# แปลงมุมเป็นเรเดียน
theta_rad = np.radians(theta)
```

```python
# เวลาบินสูงสุด
t_flight = 2 * v0 * np.sin(theta_rad) / g
```

```python
# สร้างช่วงเวลา
t = np.linspace(0, t_flight, num=100)
```

```python
# คำนวณตำแหน่ง
x = v0 * np.cos(theta_rad) * t
y = v0 * np.sin(theta_rad) * t - 0.5 * g * t**2
```

```python
# Plot กราฟ
plt.plot(x, y, label=f"v0={v0} m/s, θ={theta}°")
plt.title("Projectile Motion Simulation")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.grid(True)
plt.legend()
plt.show()
```
